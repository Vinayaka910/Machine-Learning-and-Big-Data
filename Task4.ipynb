```python

# TASK 4: BIG DATA PROCESSING & OPTIMISATION

import os
import time

# Forcing Spark to use Python 3.11

os.environ["PYSPARK_PYTHON"] = r"C:\Users\admin\AppData\Local\Programs\Python\Python311\python.exe"
os.environ["PYSPARK_DRIVER_PYTHON"] = r"C:\Users\admin\AppData\Local\Programs\Python\Python311\python.exe"

from pyspark.sql import SparkSession
from pyspark import StorageLevel


# Creation of Spark Session


spark = SparkSession.builder \
    .appName("Task4_BigData_Optimisation") \
    .master("local[*]") \
    .getOrCreate()


print("SPARK SESSION CREATED")



# Loading the Dataset


df = spark.read.csv(
    "DelayData.csv",
    header=True,
    inferSchema=True
)

print("\nDataset Loaded Successfully")

print("Rows:", df.count())
print("Columns:", len(df.columns))


# Resource Configuration


print("\nRESOURCE CONFIGURATION")

print("Application Name:")
print(spark.sparkContext.appName)

print("\nSpark Version:")
print(spark.version)

print("\nDefault Parallelism:")
print(spark.sparkContext.defaultParallelism)


# Current Partitions


print("\nCURRENT PARTITIONS")


print(df.rdd.getNumPartitions())


# Repartitioning Strategy


df_repartitioned = df.repartition(8)

print("\nPARTITIONS AFTER REPARTITION")
print(df_repartitioned.rdd.getNumPartitions())




df_repartitioned.cache()

print("\nCACHE STATUS")

print(df_repartitioned.is_cached)


# Persist Dataset


df_repartitioned.persist(
    StorageLevel.MEMORY_AND_DISK
)

print("\nSTORAGE LEVEL")

print(df_repartitioned.storageLevel)


# Performance Before Cache


start = time.time()

df.groupBy("year").count().show()

end = time.time()

print(
    "Execution Time Before Cache:",
    round(end - start, 2),
    "seconds"
)


# Performance After Cache


start = time.time()

df_repartitioned.groupBy("year").count().show()

end = time.time()

print(
    "Execution Time After Cache:",
    round(end - start, 2),
    "seconds"
)


# Generating the Spark Jobs


print("\nGENERATING SPARK JOBS")


df_repartitioned.count()


# Spark UI Information

print("\nSPARK UI")

print("Open the following URL in your browser:")
print("http://localhost:4040")

print("\nSpark Internal UI URL:")
print(spark.sparkContext.uiWebUrl)

# Keeping Session Alive for Spark UI

import time

print("\nSpark Session will remain active for 10 minutes.")
print("Take all Spark UI screenshots during this time.")

time.sleep(600)

# Stopping the Spark Session

spark.stop()

print("\nSpark Session Stopped Successfully")
```
