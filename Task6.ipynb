```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import hour
```

```python
spark = SparkSession.builder \
    .appName("Task6_Tableau_Visualisation") \
    .getOrCreate()

print("Spark Session Created")
```

```python
df = spark.read.csv(
    "/content/DelayData.csv",
    header=True,
    inferSchema=True
)
```

```python

df.show(5)
```

```python
df.write.mode("overwrite") \
    .option("header", True) \
    .csv("DelayData_Tableau")

print("Tableau Export Completed")
```
